```python
# =============================================================
# TASK 5 - CREDIT CARD FRAUD DETECTION
# CodSoft Data Science Internship
# Dataset: creditcard.csv
# =============================================================

# ── STEP 1: Upload & Load Dataset ────────────────────────────
from google.colab import files
import io, warnings
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
warnings.filterwarnings("ignore")

uploaded = files.upload()
filename = list(uploaded.keys())[0]
df = pd.read_csv(io.BytesIO(uploaded[filename]))
print("✅ Dataset loaded:", df.shape)
print(df.head())


# ── STEP 2: EDA ───────────────────────────────────────────────
print("\n📋 Missing Values:")
print(df.isnull().sum().sum(), "missing values")

print("\n⚖️ Class Distribution:")
print(df['Class'].value_counts())
print(f"\n   Fraud %    : {df['Class'].mean()*100:.4f}%")
print(f"   Non-Fraud %: {(1 - df['Class'].mean())*100:.4f}%")

fig, axes = plt.subplots(1, 2, figsize=(14, 5))
fig.suptitle('Credit Card Fraud EDA', fontsize=16, fontweight='bold')

# Class Distribution
sns.countplot(x='Class', data=df, palette='Set2', ax=axes[0])
axes[0].set_title('Class Distribution (0=Legit, 1=Fraud)')
axes[0].set_xticklabels(['Legit', 'Fraud'])

# Transaction Amount Distribution
df.groupby('Class')['Amount'].plot(kind='hist', bins=50, alpha=0.6,
                                    ax=axes[1], legend=True)
axes[1].set_title('Transaction Amount by Class')
axes[1].set_xlabel('Amount')
axes[1].legend(['Legit', 'Fraud'])

plt.tight_layout()
plt.savefig('eda_plots.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ EDA plots saved!")


# ── STEP 3: Preprocessing ────────────────────────────────────
from sklearn.preprocessing import StandardScaler

df_model = df.copy()

# Normalize Amount and Time
scaler = StandardScaler()
df_model['Amount'] = scaler.fit_transform(df_model[['Amount']])
df_model['Time']   = scaler.fit_transform(df_model[['Time']])

X = df_model.drop(columns=['Class'])
y = df_model['Class']

print("\n✅ Preprocessing complete!")
print(f"   Features: {X.shape[1]} | Samples: {X.shape[0]}")


# ── STEP 4: Handle Class Imbalance with Undersampling ────────
fraud     = df_model[df_model['Class'] == 1]
non_fraud = df_model[df_model['Class'] == 0].sample(n=len(fraud)*3,
                                                      random_state=42)
balanced  = pd.concat([fraud, non_fraud]).sample(frac=1, random_state=42)

X_bal = balanced.drop(columns=['Class'])
y_bal = balanced['Class']

print(f"\n⚖️ Balanced Dataset:")
print(f"   Fraud: {y_bal.sum()} | Non-Fraud: {(y_bal == 0).sum()}")


# ── STEP 5: Train / Test Split ───────────────────────────────
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(
    X_bal, y_bal, test_size=0.2, random_state=42, stratify=y_bal
)
print(f"\n📐 Train: {X_train.shape[0]} rows | Test: {X_test.shape[0]} rows")


# ── STEP 6: Train Models ──────────────────────────────────────
from sklearn.linear_model import LogisticRegression
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import (accuracy_score, classification_report,
                              confusion_matrix, ConfusionMatrixDisplay)

# Logistic Regression
lr = LogisticRegression(max_iter=1000, random_state=42)
lr.fit(X_train, y_train)
lr_preds = lr.predict(X_test)
lr_acc   = accuracy_score(y_test, lr_preds)

# Random Forest
rf = RandomForestClassifier(n_estimators=100, random_state=42, n_jobs=-1)
rf.fit(X_train, y_train)
rf_preds = rf.predict(X_test)
rf_acc   = accuracy_score(y_test, rf_preds)

print(f"\n🔵 Logistic Regression Accuracy : {lr_acc*100:.2f}%")
print(f"🟢 Random Forest Accuracy       : {rf_acc*100:.2f}%")


# ── STEP 7: Evaluate Best Model ──────────────────────────────
print("\n📊 Classification Report (Random Forest):")
print(classification_report(y_test, rf_preds,
      target_names=['Legit', 'Fraud']))

cm = confusion_matrix(y_test, rf_preds)
disp = ConfusionMatrixDisplay(cm, display_labels=['Legit', 'Fraud'])
fig, ax = plt.subplots(figsize=(6, 5))
disp.plot(ax=ax, colorbar=False, cmap='Blues')
ax.set_title('Confusion Matrix — Random Forest')
plt.tight_layout()
plt.savefig('confusion_matrix.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Confusion matrix saved!")


# ── STEP 8: Feature Importance ───────────────────────────────
feat_imp = pd.Series(rf.feature_importances_,
                     index=X_bal.columns).sort_values(ascending=False).head(10)
fig, ax = plt.subplots(figsize=(9, 5))
feat_imp.sort_values().plot(kind='barh', color='steelblue', ax=ax)
ax.set_title('Top 10 Feature Importances — Random Forest')
ax.set_xlabel('Importance Score')
plt.tight_layout()
plt.savefig('feature_importance.png', dpi=150, bbox_inches='tight')
plt.show()
print("✅ Feature importance saved!")

```
